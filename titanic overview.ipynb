
##  Children who died , survived with/without parch (children age-15)

```python
children=df[df['Age']<=15]
children
```

```python
orphans=children[children['Parch']==0]
orphans
```

[SEE THIS LINK TO UNDERSTAND BELOW BLOCK OF CODE](https://stackoverflow.com/questions/13770394/make-output-cells-like-markdown)

```python
htmlContent = ''

def header(text):
    raw_html = '<h3>' + text + '</h3>'
    return raw_html

def box(text):
    raw_html = '<div style="border:1px dotted black;padding:2em;">'+str(text)+'</div>'
    return raw_html

def addContent(raw_html):
    global htmlContent
    htmlContent += raw_html

a=1
# Example
addContent( header("Children and Orphans") )
addContent( box(f"out of {children.shape[0]} children, {orphans.shape[0]} are orphans") )

from IPython.core.display import HTML
HTML(htmlContent)
```

```python
STR=f"{orphans[orphans['Survived']==1].shape[0]} no.of orphans {orphans[(orphans['Survived']==1) & (orphans['Sex']=='male')].shape[0]} male and {orphans[(orphans['Survived']==1) & (orphans['Sex']=='female')].shape[0]} female survived and {orphans[orphans['Survived']==0].shape[0]} no.of orphans {orphans[(orphans['Survived']==0) & (orphans['Sex']=='male')].shape[0]} male and {orphans[(orphans['Survived']==0) & (orphans['Sex']=='female')].shape[0]} female, died out of a total of {orphans.shape[0]} number of orphans"

from IPython.core.display import HTML
html_of_str='<h3>'+STR+'</h3>'

HTML(html_of_str)
```
